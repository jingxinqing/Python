```python
import json

def show_growth(l):
    if l[0]['region'] == 'United States of America':
        print('USA Population Growth')
    if l[0]['region'] == 'Russian Federation':
        print('Russian Population Growth')
    first_item = l[0]
    last_item = l[-1]
    idx = 0
    for item in l:
        current_population = item['population']
        #print(current_population)
        if idx == 0:
            previous_population = l[idx]['population']
        else:
            previous_population = l[idx-1]['population']
        #print(previous_population)   
        growth_rate = ((current_population - previous_population)/previous_population)*100
        print('   ', item['year'], item['population'], "{:.3f}".format(growth_rate))
        idx = idx + 1
    overall_growth = ((last_item['population']-first_item['population'])/first_item['population'])*100
    print('Overall growth:', "{:.2f}".format(overall_growth))

    
f = open("population.json", 'r')
population_list = json.load(f)
#print(population_list[:2])

usa_population = []
russia_population = []
for item in population_list:
    if item['region'] == 'United States of America' and item['year'] > '1979':
        usa_population.append(item)
    if item['region'] == 'Russian Federation' and item['year'] > '1979':
        russia_population.append(item)
#print(usa_population[:2])
#print(russia_population[:2])

show_growth(usa_population)
show_growth(russia_population)   
```
